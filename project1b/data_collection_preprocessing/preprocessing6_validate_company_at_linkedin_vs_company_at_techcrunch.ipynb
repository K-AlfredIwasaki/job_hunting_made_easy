```python
import pandas as pd
import re
data = pd.read_csv("article_after_processing4.csv", encoding='iso-8859-1')
```

```python
data
```

```python
def get_linkedin_company_name(link):
    if re.search(r'^https://www\.linkedin\.com/company/', link):
        return re.sub(r'^https://www\.linkedin\.com/company/', '', link).replace("-", " ")
    else:
        return "error"
    
data["Company_at_Linkedin"] = data.Linkedin_profile.apply(
    lambda company: get_linkedin_company_name(company) )
```
