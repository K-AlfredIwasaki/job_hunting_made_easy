```python
def write_company_info(company_list):
    import csv
    csvFile = open("linkedin_link_list.csv", 'a', newline='', encoding="utf8")
    try:
        writer = csv.writer(csvFile)
        for company in company_list:
            link = get_link(company)
            writer.writerow( (
                company,
                link
            ) )
    finally:
        csvFile.close()
```
