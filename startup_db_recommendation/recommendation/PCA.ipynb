```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("article_after_processing10.csv", encoding='iso-8859-1')
```

```python
industry_attributes = data.loc[:,"spc_Logistics and Supply Chain": "Real Estate"]
```

```python
industry_attributes.shape
```

```python
pca = PCA(n_components=30)
pca.fit(industry_attributes)
```

```python
np.round(pca.explained_variance_ratio_, decimals = 4) * 100
```

```python
# cumulative
var1 = np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) *100)
var1
```

```python
%matplotlib inline
plt.plot(var1)
```

```python
pca = PCA(n_components=10)
pca.fit(industry_attributes)
df = pd.DataFrame(pca.transform(industry_attributes))
df
```
