```python
# import get_company_from_title function
from company_from_title import *
```

```python
import pandas as pd
import re
data = pd.read_csv("article_after_processing1.csv", encoding='iso-8859-1')
```

```python
data["Company"] = data.title.apply(
    lambda item: get_company_from_title(item))
```

```python
# drop duplicate rows
data.drop_duplicates(subset = ["title"], inplace = True)
data.drop_duplicates(subset = ["excerpt"], inplace = True)
```

```python
# check the rows with long company names
# but don't take any action for now
mask = data.Company.apply(lambda company: 
                          True if count_word(company) > 3 else False)
data.loc[mask][["title", "Company"]].shape
```

```python
# remove " s" in "s Company Name"
data.Company = data.Company.apply(lambda company:
                                 re.sub(r'^s\s|^\s', '', company))
```
