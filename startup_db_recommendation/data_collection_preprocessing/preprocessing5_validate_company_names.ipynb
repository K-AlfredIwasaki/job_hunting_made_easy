```python
import pandas as pd
import re
data = pd.read_csv("article_after_processing4.csv", encoding='iso-8859-1')
```

```python
def get_linkedin_company_name(link):
    if isinstance(link, str) and re.search(
        r'^https://www\.linkedin\.com/company/', link):
        return re.sub(r'^https://www\.linkedin\.com/company/',
                      '', link).replace("-", " ")
    else:
        pass

data["Company_at_Linkedin"] = data.linkedin_link.apply(
    lambda company: get_linkedin_company_name(company) )
```

```python
# further clean up
def clean_up_linked_in_link(link):
    regex = r'^https://www\.linkedin\.com/company/'
    regex2 = r'(https:\/\/www\.linkedin\.com\/company\/\w+-*\w*-*\w*-*\w*-*\w*)[&/%]'
    if isinstance(link, str) and re.search(regex, link):
        return link
    elif isinstance(link, str) and re.search(regex2,link):
        link = re.search(regex2,link)
        return link.group(1)
    else:
        return link
    
data["linkedin_link"] = data.linkedin_link.apply(
    lambda link: clean_up_linked_in_link(link) )


data["Company_at_Linkedin"] = data.linkedin_link.apply(
    lambda company: get_linkedin_company_name(company) )
```

```python
def check_company(row):
    regex = re.escape(row["Company"].split(" ")[0])
    if re.search(regex, row["Company_at_Linkedin"],re.IGNORECASE):
        return True
    else:
        return False
    
data["company_check"] = data.apply(lambda row: check_company(row), axis = 1)
```
